## Importing data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/monitoramento_tempo.csv')
```

```python
data.head()
```

## changing variables for english

```python
data.rename(columns = {
    'temperatura': 'temperature',
    'pressão': 'pressure',
    'humidade': 'humidity',
    'direção do vento': 'wind_direction',
    'velocidade do vento': 'wind_speed',
    'dia_da_semana': 'weekday',
    'data': 'datetime'
}, inplace= True)
```

```python
data.head()
```

## analyzing if the variables have null values

```python
data.info()
```

## changing datetime variable Type

```python
data['datetime'] = pd.to_datetime(data['datetime'])
```

```python
data.info()
```

## <font color='red' style='font-size: 30px;'>1.3 Creating figure</font>
<hr style='border: 2px solid red;'>

```python
fig, ax = plt.subplots()

fig.set_size_inches(15, 8)
ax.plot(data['datetime'], data['temperature'])

plt.title('Temperature in Time', fontsize = 20, y = 1.05)
ax.set_xlabel('Time - Years', fontsize = 16)
ax.set_ylabel('Temperature - Kevin', fontsize = 16)
```

## Setting figure

```python
#creating figure
fig = plt.figure(figsize = (15, 8))

# creating axis and renaming labels
axis = fig.add_axes([0, 0, 1, 1])
axis.plot(data['datetime'], data['temperature'], color = 'skyblue')
axis.set_title('Temperature in moment', fontsize = 20, y = 1.05)
axis.set_ylabel('Temperature - Kevin', fontsize = 16)
axis.set_xlabel('Time - Years', fontsize = 16)

# create legend
axis.legend(['temperature'], loc = 'lower right', fontsize = 14)
```
