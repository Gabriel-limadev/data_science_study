## Importing data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/monitoramento_tempo.csv')
```

```python
data.head()
```

## changing variables for english

```python
data.rename(columns = {
    'temperatura': 'temperature',
    'pressão': 'pressure',
    'humidade': 'humidity',
    'direção do vento': 'wind_direction',
    'velocidade do vento': 'wind_speed',
    'dia_da_semana': 'weekday',
    'data': 'datetime'
}, inplace= True)
```

```python
data.head()
```

## analyzing if the variables have null values

```python
data.info()
```

## changing datetime variable Type

```python
data['datetime'] = pd.to_datetime(data['datetime'])
```

```python
data.info()
```

## <font color='red' style='font-size: 30px;'>1.3 Creating figure</font>
<hr style='border: 2px solid red;'>

```python
fig, ax = plt.subplots()

fig.set_size_inches(15, 8)
ax.plot(data['datetime'], data['temperature'])

ax.set_title('Temperature in Time', fontsize = 20, y = 1.05)
ax.set_xlabel('Time - Years', fontsize = 16)
ax.set_ylabel('Temperature - Kevin', fontsize = 16)
```

## Setting figure

```python
#creating figure
fig = plt.figure(figsize = (15, 8))

# creating axis and renaming labels
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(data['datetime'], data['temperature'], color = 'skyblue')
ax.set_title('Temperature in moment', fontsize = 20, y = 1.05)
ax.set_ylabel('Temperature - Kevin', fontsize = 16)
ax.set_xlabel('Time - Years', fontsize = 16)

# create legend
ax.legend(['temperature'], loc = 'lower right', fontsize = 14)
```

## line-styles, grid and more

```python
from datetime import datetime
```

```python
#creating figure
fig = plt.figure(figsize = (15, 8))

# creating axis
ax = fig.add_axes([0, 0, 1, 1])

# Line Styles::
    # line-width -> lw = 2
    # line-style -> 'dotted'
    # marker -> marker = 'o'
ax.plot(data['datetime'], data['temperature'], color = 'skyblue')

# delimiting x axis
ax.set_xlim(datetime(2014, 1, 1), datetime(2015, 1, 1))

# Renaming labels
ax.set_title('Temperature in moment', fontsize = 20, y = 1.05)
ax.set_ylabel('Temperature - Kevin', fontsize = 16)
ax.set_xlabel('Time - 2014', fontsize = 16)

# create legend
ax.legend(['temperature'], loc = 'lower right', fontsize = 14)

ax.grid(True)
```

## Figure inside figure

```python
# Creating a figure for put the axies
fig = plt.figure(figsize = (15, 8))

# Creating two axies
ax = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

# Setting first axis
    # xlim or ylim are the form for setting axis wight
ax.set_xlim(datetime(2014, 1, 1), datetime(2015, 1, 1))
ax.plot(data['datetime'], data['temperature'], color = 'skyblue')
ax.set_title('Temperature in 2014', fontsize = 20, y = 1.05)
ax.set_xlabel('Time - 2014', fontsize = 16)
ax.set_ylabel('Temperature - Kevin', fontsize = 16)
ax.legend(['Temperature'], loc = 'lower right', fontsize = 14)
ax.grid(True)

# Setting second axis
ax2.plot(data['datetime'], data['temperature'], color  = 'slategray')
ax2.set_title('Temperature in Moment', fontsize = 16)
ax2.set_xlabel('Time - 2014', fontsize = 12)
ax2.set_ylabel('Temperature - Kevin', fontsize = 12)
ax2.legend(['Temperature'], loc = 'best', fontsize = 10)
ax2.grid(True)
```

## Focusing in month 

```python
fig = plt.figure(figsize = (15, 8))

ax = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

ax.set_xlim(datetime(2014, 5, 1), datetime(2014, 6, 1))
ax.set_ylim(270, 320)
ax.plot(data['datetime'], data['temperature'], color = 'skyblue')
ax.set_title('Temperature in May 2014', fontsize = 20, y = 1.05)
ax.set_xlabel('Time - 2014', fontsize = 16)
ax.set_ylabel('Temperature - Kelvin', fontsize = 16)
ax.legend(['Temperature'], loc = 'lower right', fontsize = 14)
ax.grid(True)

# Creating filter
    # Here i am separing the data that are before and after may
gray_left = data['datetime'] < datetime(2014, 5, 1)
gray_right = data['datetime'] > datetime(2014, 6, 1)

ax2.set_xlim(datetime(2014, 1, 1), datetime(2015, 1, 1))
ax2.plot(data['datetime'], data['temperature'], color  = 'skyblue')

    # Here i set the color gray  for all data that gray_left = data['datetime'] < datetime(201
ax2.plot(data[gray_left]['datetime'], data[gray_left]['temperature'], color = 'slategray')
ax2.plot(data[gray_right]['datetime'], data[gray_right]['temperature'], color = 'slategray')

ax2.set_title('Temperature in 2014', fontsize = 16)
ax2.set_xlabel('Time - 2014', fontsize = 12)
ax2.set_ylabel('Temperature - Kelvin', fontsize = 12)
ax2.legend(['Temperature'], loc = 'best', fontsize = 10)
ax2.grid(True)
```

## <font color='red' style='font-size: 30px;'>1.5 Notes and more Visualizations</font>
<hr style='border: 2px solid red;'>

```python
fig = plt.figure(figsize = (15, 8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(data['datetime'], data['temperature'], color = 'skyblue')
ax.set_title('Temperature in moment', fontsize = 20, y = 1.05)
ax.set_xlabel('Time - Years', fontsize = 16)
ax.set_ylabel('Temperature - Kelvin', fontsize = 16)
ax.legend(['Temperature'], fontsize = 14)
ax.grid(True)


# ---------------------------------- Creating Annotate ---------------------------------------
# ------------------------ Maximum
x1 = data['datetime'][data['temperature'].idxmax()] # - Getting the time where it has the highest temperature
y1 = max(data['temperature']) # Getting the highest temperature

# Here x2 and y2 will be the beginning, while x1 and y2 will be arrow final
x2 = data['datetime'][data['temperature'].idxmax() - 7000]
y2 = max(data['temperature']) - 5

ax.annotate(
    'Maximum', xy = (x1, y1), 
    fontsize = 16, 
    xytext = (x2, y2), # the text "Maximum" will be this position
    arrowprops = dict(facecolor='k') # here i am doing the arrow
)

# ------------------------ Minimum
x1_min = data['datetime'][data['temperature'].idxmin()] # - Getting the time where it has the lowest temperature
y1_min = min(data['temperature']) # Getting the lowest temperature

x2_min = data['datetime'][data['temperature'].idxmin() - 7000]
y2_min = min(data['temperature']) + 5
ax.annotate(
    'Minimum',
    xy = (x1_min, y1_min),
    fontsize = 16,
    xytext = (x2_min, y2_min), # the text "Minimum" will be this position
    arrowprops = dict(facecolor='k') # here i am doing the arrow
)



# Creating marking
    ## k -> 'black'
    ## ls -> linestyles
    
ax.axhline(max(data['temperature']), color = 'k', ls = '--')
ax.axhline(min(data['temperature']), color = 'k', ls = '--')
```

## More Visualizations

```python
temperature_per_weekday = data.groupby('weekday')['temperature'].mean()
days_name = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperature_per_weekday = temperature_per_weekday[days_name]
temperature_per_weekday
```

```python
fig = plt.figure(figsize = (5, 4))
ax = fig.add_axes([0, 0, 1, 1])

# creating bar grafhic with index
index = range(len(temperature_per_weekday))
colors = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']
ax.bar(index, temperature_per_weekday, color = colors, edgecolor = 'k')

# Settings axies names
ax.set_title('Temperature per WeekDay', fontsize = 20, y = 1.05)
ax.set_xlabel('WeekDay', fontsize = 16)
ax.set_ylabel('Mean Temperature', fontsize = 16)

# Changing index
ax.set_xticks(index)
ax.set_xticklabels(days_name);
```

## pizza chart

```python
fig = plt.figure(figsize = (5, 4))
ax = fig.add_axes([0, 0, 1, 1])

explode = [0.1, 0, 0, 0, 0, 0, 0.1]

ax.pie(
    temperature_per_weekday, 
    labels = temperature_per_weekday.index,
    autopct = '%.1f%%',
    explode = explode,
    shadow = True

)
ax.set_title('Temperature per weekday', fontsize = 16, y = 1.05)
```
