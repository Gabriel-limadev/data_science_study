## Importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/tmdb_5000_movies.csv')
data.head()
```

```python
data.describe().round(2)
```

## Analysing vote average with graphic

### Matplotlib

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (16, 8))

ax.hist(data['vote_average'], bins = 80, color ='skyblue')
ax.set_title('Frequency Distribution - Vote Average', fontsize = 20, y = 1.05)
ax.set_xlabel('Vote Average', fontsize = 16)
ax.set_ylabel('Frequency', fontsize = 16)
ax.grid(True)
ax.tick_params(labelsize = 12);
```

### Seaborn

```python
ax = sns.histplot(data['vote_average'], color = 'skyblue', kde = True, bins = 80)
ax.figure.set_size_inches(16, 8)
ax.set_title('Frequency Distribution - Vote Average', fontsize = 20)
ax.set_xlabel('Vote Average', fontsize = 16)
ax.set_ylabel('Frequency', fontsize = 16)
ax.grid(True)
ax.tick_params(labelsize = 12);
```

## Boxplot

### Matplotlib

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (16, 8))

ax.boxplot(data['vote_average'], vert = False, patch_artist = True)


# Case if i wanted change the color, i would have that run code below: (Besides attribute the line above for a variable)
# for box in boxes['boxes']:
#     box.set_color('skyblue')

ax.set_title('Box plot - Vote Average', fontsize = 20, y = 1.05)
ax.set_xlabel('Vote Average', fontsize = 16);
```

### Seaborn

```python
ax = sns.boxplot(
    x = 'vote_average', 
    data = data,
    orient = 'h',
    width = 0.2
)
ax.figure.set_size_inches(16, 8)
ax.set_title('Box plot - Vote Average', fontsize = 20, y = 1.05)
ax.set_xlabel('Vote Average', fontsize = 16);
```

## Some movies had less that 10 votes, i removed them

```python
tmdb_with_more_9_votes = data.query('vote_count >= 10')
```

```python
tmdb_with_more_9_votes.describe()
```

```python
ax = sns.histplot(tmdb_with_more_9_votes['vote_average'], color = 'skyblue', kde = True)
ax.figure.set_size_inches(16, 8)
ax.set_title('Frequency Distribution - Vote Average', fontsize = 20, y = 1.05)
ax.set_xlabel('Movies with vote average more than 9 votes', fontsize = 16)
ax.set_ylabel('Frequency', fontsize = 16)
ax.grid(True)
ax.tick_params(labelsize = 12);
```

## Cumulative distribution function

```python
ax = sns.histplot(
    tmdb_with_more_9_votes['vote_average'], 
    color = 'skyblue', 
    kde = True,
    stat="density", 
    common_norm=False,
    cumulative = True,
    kde_kws = {
        'cumulative': True
    }
)
ax.lines[0].set_color('k')
ax.figure.set_size_inches(16, 8)
ax.set_title('Frequency Distribution - Vote Average', fontsize = 20, y = 1.05)
ax.set_xlabel('Movies with vote average more than 9 votes', fontsize = 16)
ax.set_ylabel('Cumulative percentage of movies', fontsize = 16)
ax.grid(True)
ax.tick_params(labelsize = 12);
```

## Box plot

```python
ax = sns.boxplot(
    x = 'vote_average', 
    data = tmdb_with_more_9_votes,
    orient = 'h',
    width = 0.2
)
ax.figure.set_size_inches(16, 8)
ax.set_title('Box plot - Vote Average', fontsize = 20, y = 1.05)
ax.set_xlabel('Movies with vote average more than 9 votes', fontsize = 16);
```

## <font color='red' style='font-size: 30px;'>1.3 Analysing the Movielens</font>
<hr style='border: 2px solid red;'>

```python
ratings = pd.read_csv('../data/ratings.csv')
ratings
```

```python
note_average_per_movie = ratings.groupby('movieId').mean()['rating']
note_average_per_movie.head()
```

```python
ax = sns.histplot(note_average_per_movie.values, color = 'skyblue', kde = True)
ax.figure.set_size_inches(16, 8)
ax.set_title('Frequency Distribution - Note Average in MovieLens', fontsize = 20, y = 1.05)
ax.set_xlabel('Note Average per Movie', fontsize = 16)
ax.set_ylabel('Frequency', fontsize = 16)
ax.grid(True)
ax.tick_params(labelsize = 12);
```

```python
movies_with_more_than_9_votes.valuesnumber_votes_per_movie = ratings.groupby('movieId').count()
number_votes_per_movie
```

```python
movies_with_more_than_9_votes = number_votes_per_movie.query('rating >= 10').index
movies_with_more_than_9_votes.values
```

```python
note_average_movies_with_more_than_9_votes = note_average_per_movie.loc[movies_with_more_than_9_votes.values]
note_average_movies_with_more_than_9_votes
```

```python
ax = sns.histplot(note_average_movies_with_more_than_9_votes, color = 'skyblue', kde = True)
ax.figure.set_size_inches(16, 8)
ax.set_title('MovieLens - Note Average with more than 9 votes', fontsize = 20, y = 1.05)
ax.set_xlabel('Note Average per Movie', fontsize = 16)
ax.set_ylabel('Frequency', fontsize = 16)
ax.grid(True)
ax.tick_params(labelsize = 12);
```

```python
ax = sns.boxplot(
    x = note_average_movies_with_more_than_9_votes,
    orient = 'h',
    width = 0.2
)
ax.figure.set_size_inches(16, 8)
ax.set_title('MovieLens - Vote Average', fontsize = 20, y = 1.05)
ax.set_xlabel('Movies with vote average more than 9 votes', fontsize = 16);
```

## Cumulative distribution function

```python
ax = sns.histplot(
    note_average_movies_with_more_than_9_votes, 
    color = 'skyblue', 
    kde = True,
    stat="density", 
    common_norm=False,
    cumulative = True,
    kde_kws = {
        'cumulative': True
    }
)
ax.lines[0].set_color('k')
ax.figure.set_size_inches(16, 8)
ax.set_title('MovieLens - Note Average with more than 9 votes', fontsize = 20, y = 1.05)
ax.set_xlabel('Note Average per Movie', fontsize = 16)
ax.set_ylabel('Cumulative percentage of movies', fontsize = 16)
ax.grid(True)
ax.tick_params(labelsize = 12);
```
