```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 8))
```

```python
data = pd.read_csv('data/aluguel.csv', sep = ';')
data.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(data['Valor'], data['Area'])
g1.set_title('Valor x Area')

g2.hist(data['Valor'])
g2.set_title('Histograma de Valor')

data_g3 = data['Valor'].sample(100)
data_g3.index = range(data_g3.shape[0])
g3.plot(data_g3)
g3.set_title('Amostra valor')

group = data.groupby('Tipo')['Valor']
label = group.mean().index
values = group.mean().values
g4.bar(label, values)
g4.set_title('Valor Média por tipo')
       
```

```python
area = plt.figure()
g1 = area.add_subplot(1, 2, 1)
g2 = area.add_subplot(1, 2, 2)
grupo1 = data.groupby('Tipo Agregado')['Valor']
label = grupo1.count().index
valores = grupo1.count().values
g1.pie(valores, labels = label, autopct='%1.1f%%')
g1.set_title('Total de Imóveis por Tipo Agregado')
grupo2 = data.groupby('Tipo')['Valor']
label = grupo2.count().index
valores = grupo2.count().values
g2.pie(valores, labels = label, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2.set_title('Total de Imóveis por Tipo')
```
