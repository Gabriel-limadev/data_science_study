# Creating new variables


```python
import pandas as pd

data = pd.read_csv('data/aluguel_residencial.csv', sep = ';')

data.head()
```

## Bruto Value

```python
data['Valor Bruto'] = data['Valor'] + data['Condominio'] + data['IPTU']
```

```python
data.head()
```

## Value per square meter 

```python
data['m2 Value'] = data['Valor'] / data['Area']
data['Valor m2'] = data['Valor'] / data['Area']
data.head()
```

## Deleting column

```python
data.drop(['m2 Value'], axis = 1, inplace = True)
```

## Setting value after comma

```python
# Geral 

pd.set_option('display.precision', 2)

# Especifico
data['Valor m2'] = data['Valor'].round(2)

data.head(10)
```

## Valor m2 Bruto


```python
data['Valor Bruto m2'] = (data['Valor Bruto'] / data['Area']).round(2)
data.head()
```

## Organizing Tipo

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']

# Here i am creating a new variable called "Tipo Agregado".
# In it was give a function lambda that fill the field with "Casa" case "Tipo" be "Casa" else
# fill "Apartamento"

data['Tipo Agregado'] = data['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
data
```

## Deleting variables - columns p2

```python
dados_aux = pd.DataFrame(data[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])

dados_aux.head()
```

### del

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux.head()
```

### .pop

```python
dados_aux.pop('Valor Bruto m2')
```

### .drop

```python
data.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
```

## Exporting

```python
data.to_csv('data/aluguel_residencial.csv', sep = ';', index = False)
```
